```python
import pandas as pd
```

```python
df1 = pd.read_csv('IDEFC_NM_sep24.csv', encoding='ISO-8859-1')
```

```python
df1
```

```python
df2 = pd.read_csv('IDEFC_ago2024.csv', encoding='ISO-8859-1')
```

```python
df2 = df2[df2['AÑO'] < 2015]
df2
```

```python
columnas_df1 = df1.columns
columnas_df1
```

```python
columnas_df2 = df2.columns
columnas_df2
```

```python
df1 = df1.drop(columns=[columnas_df1[1]])
columnas_df1 = df1.columns
columnas_df1
```

```python
selected_columns = [columnas_df1[2], columnas_df1[3], columnas_df1[4], columnas_df1[5]]
combinaciones_df1 = df1[selected_columns].drop_duplicates()
combinaciones_df1
```

```python
lines_to_remove = [22, 32, 33, 34, 35, 36, 37, 38, 75, 76, 77, 80, 81, 82, 83, 84, 89, 91, 92, 93, 94, 95, 96, 97]
combinaciones_df1 = combinaciones_df1.drop(index=lines_to_remove)
combinaciones_df1
```

```python
df1_delitos_selec = df1.merge(combinaciones_df1,
                              on=selected_columns,
                              how='inner')
df1_delitos_selec
```

```python
combinaciones_df1_v2 = df1_delitos_selec[selected_columns].drop_duplicates()
combinaciones_df1_v2
```

```python
df2 = df2.drop(columns=[columnas_df2[1]])
columnas_df2 = df2.columns
columnas_df2
```

```python
selected_columns2 = [columnas_df2[2], columnas_df2[3], columnas_df2[4]]
combinaciones_df2 = df2[selected_columns2].drop_duplicates()
combinaciones_df2
```

```python
lines_to_remove2 = [0, 2, 3, 7, 17, 18, 19, 20]
combinaciones_df2 = combinaciones_df2.drop(index=lines_to_remove2)
combinaciones_df2
```

```python
df2_delitos_selec = df2.merge(combinaciones_df2,
                              on=selected_columns2,
                              how='inner')
df2_delitos_selec
```

```python
combinaciones_df2_v2 = df2_delitos_selec[selected_columns2].drop_duplicates()
combinaciones_df2_v2
```

```python
df1_delitos = df1_delitos_selec.drop(columns=[columnas_df1[2], columnas_df1[3], columnas_df1[4], columnas_df1[5]])
df1_delitos.columns = df1_delitos.columns.str.lower()
columnas_df1 = df1_delitos.columns
columnas_df1
```

```python
id_vars1 = [columnas_df1[0], columnas_df1[1]]
value_vars1 = [columnas_df1[2], columnas_df1[3], columnas_df1[4], columnas_df1[5], columnas_df1[6], columnas_df1[7], columnas_df1[8], columnas_df1[9], columnas_df1[10], columnas_df1[11], columnas_df1[12], columnas_df1[13]]
```

```python
df1_melted = pd.melt(df1_delitos, id_vars=id_vars1,
                    value_vars=value_vars1,
                    var_name='mes', value_name='eventos').dropna()
df1_melted
```

```python
df1_final = df1_melted.groupby(['año', 'entidad', 'mes'], as_index=False)['eventos'].sum()
df1_final
```

```python
df2_delitos = df2_delitos_selec.drop(columns=[columnas_df2[2], columnas_df2[3], columnas_df2[4]])
df2_delitos.columns = df2_delitos.columns.str.lower()
columnas_df2 = df2_delitos.columns
columnas_df2
```

```python
df2_melted = pd.melt(df2_delitos, id_vars=id_vars1,
                    value_vars=value_vars1,
                    var_name='mes', value_name='eventos').dropna()
df2_melted
```

```python
df2_final = df2_melted.groupby(['año', 'entidad', 'mes'], as_index=False)['eventos'].sum()
df2_final
```

```python
df1_final['entidad'] = df1_final['entidad'].str.lower()
df2_final['entidad'] = df2_final['entidad'].str.lower()
```

```python
df1_final['entidad'].unique()
```

```python
df2_final['entidad'].unique()
```

```python
df1_final['entidad'].unique() == df2_final['entidad'].unique()
```

```python
replacements = {'ciudad de méxico': 'ciudad de mexico', 
                'coahuila de zaragoza': 'coahuila', 
                'michoacán de ocampo': 'michoacan',
                'méxico': 'mexico',
                'nuevo león': 'nuevo leon',
                'querétaro': 'queretaro',
                'san luis potosí': 'san luis potosi',
                'veracruz de ignacio de la llave': 'veracruz',
                'yucatán': 'yucatan'}
```

```python
df1_final['entidad'] = df1_final['entidad'].replace(replacements)
```

```python
sorted(df1_final['entidad'].unique()) == sorted(df2_final['entidad'].unique())
```

```python
df1_final
```

```python
df2_final
```

```python
df_final = pd.concat([df1_final, df2_final], ignore_index=True)
df_final = df_final.sort_values(by=['año', 'entidad', 'mes'], ascending=[True, True, True])
df_final = df_final.reset_index(drop=True)
df_final
```

```python
df_final.to_csv('eventos_violentos.csv', index=False)
```
